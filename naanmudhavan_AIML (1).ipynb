```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
data = pd.read_csv("card.csv")
X = data.drop(['Time', 'Class'], axis=1)
model = IsolationForest(contamination=0.01, random_state=42)
model.fit(X)
outliers = model.predict(X)
outliers = pd.Series(outliers)
data['Outlier'] = outliers
plt.figure(figsize=(10,6))
plt.scatter(data[data['Outlier'] == 1].index, data[data['Outlier'] == 1]['Amount'], color='green', label='Normal')
plt.scatter(data[data['Outlier'] == -1].index, data[data['Outlier'] == -1]['Amount'], color='red', label='Fraud')
plt.xlabel('Index')
plt.ylabel('Amount')
plt.legend()
plt.title('Isolation Forest - Credit Card Fraud Detection')
plt.show()
```
